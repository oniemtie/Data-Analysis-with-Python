Handle missing data:

- For continuous attributes (e.g., age), replace missing values with the mean.
- For categorical attributes (e.g., smoker), replace missing values with the most frequent value.
- Update the data types of the respective columns.
- Verify the update using `df.info()`.


````python
# Replace missing values in 'age' with the mean
df['age'] = pd.to_numeric(df['age'], errors='coerce')  # Convert 'age' to numeric, coercing errors to NaN
mean_age = df['age'].mean()
df['age'].fillna(mean_age, inplace=True)

# Replace missing values in 'smoker' with the most frequent value
most_frequent_smoker = df['smoker'].mode()[0]
df['smoker'].fillna(most_frequent_smoker, inplace=True)

# Update the data types
df['age'] = df['age'].astype(int)
df['smoker'] = df['smoker'].astype(int)

# Verify the update using df.info()
print("\nDataframe info after handling missing data:")
df.info()
````

# Task 4 : Model Development

Fit a linear regression model that may be used to predict the `charges` value, just by using the `smoker` attribute of the dataset. Print the $ R^2 $ score of this model.


````python
X = df[['smoker']]
Y = df['charges']
lm = LinearRegression()
lm.fit(X,Y)
print(lm.score(X, Y))
````

Initialize a Ridge regressor that used hyperparameter $ \alpha = 0.1 $. Fit the model using training data data subset. Print the $ R^2 $ score for the testing data.


````python
from sklearn.linear_model import Ridge

# Initialize Ridge regressor with alpha=0.1
ridge = Ridge(alpha=0.1)

# Fit the model using training data
ridge.fit(x_train, y_train)

# Print the R-squared score for the testing data
print("R-squared score for Ridge regression on testing data:", ridge.score(x_test, y_test))
````
